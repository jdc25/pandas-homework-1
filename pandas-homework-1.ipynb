```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
pokemon_df = pd.read_csv('Pokemon.csv', sep=',')
```

```python
custom_palette = {
    'Grass': 'green',
    'Fire': 'red',
    'Water': 'blue',
    'Bug': 'orange',
    'Normal': 'slategray',
    'Electric': 'yellow',
    'Ground': 'sienna',
    'Fairy': 'pink',
    'Poison': 'purple',
    'Fighting': 'brown',
    'Psychic': 'magenta',
    'Rock': 'slategrey',
    'Ghost': 'darkviolet',
    'Ice': 'deepskyblue',
    'Dragon': 'royalblue',
    'Steel': 'slateblue',
    'Dark': 'black',
    'Flying': 'skyblue'
}

plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=pokemon_df,
    x='Attack',
    y='Defense',
    hue='Type 1',
    palette=custom_palette,
    s=30
)

plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack v. Defense')
plt.legend(title='Pokemon Type', loc='upper left', bbox_to_anchor=(1, 1))


plt.show()
```

```python
# how many Pokemon have their Attack stat above the average value for the entire set
average_attack = pokemon_df['Attack'].mean()
above_average_attack = (pokemon_df['Attack'] > average_attack).sum()
print("Pokemon with attack above average:", above_average_attack)
```
